Bias-variance tradeoff

```python
from sklearn import linear_model, neighbors
```

```python
f = lambda x: x*np.sin(x)
```

```python
xtest = np.arange(-7,7,.01).reshape(-1,1)
```

```python
n = 20
d = 8
xtrain = np.random.rand(n).reshape(-1,1)*14-7
ytrain = f(xtrain).reshape(-1) + np.random.randn(n)
ypred  = linear_model.LinearRegression().fit(
    np.hstack((xtrain**i for i in range(1,d+1))), ytrain
).predict(np.hstack((xtest**i for i in range(1,d+1))))
ytest  = f(xtest).reshape(-1)
plot(xtest,ytest,'r-')
plot(xtest,ypred,'b--')
ylim(-8,4)
```

```python
n = 20
k = 15
xtrain = np.random.rand(n).reshape(-1,1)*14-7
ytrain = f(xtrain).reshape(-1) + np.random.randn(n)
ypred  = neighbors.KNeighborsRegressor(n_neighbors=k).fit(
    xtrain, ytrain
).predict(xtest)
ytest  = f(xtest).reshape(-1)
plot(xtest,ytest,'r-')
plot(xtest,ypred,'b--')
ylim(-8,4)
```
